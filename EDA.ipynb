# Handling Outliers

```python
#loaded the Dataset because in above data we cannot find the outlier
df=sns.load_dataset('Iris')
df.columns
sns.boxplot(x=df['sepal_width'],data=df)
```

```python
#IQR
q1=np.percentile(df['sepal_width'],25,interpolation='midpoint')
q3=np.percentile(df['sepal_width'],75,interpolation='midpoint')
```

```python
IQR=q3-q1
print("old Shape",df.shape)

Upper=q3+1.5*IQR
print(Upper)

Lower=q1-1.5*IQR
print(Lower)

[Lower,Upper]
```

```python
#Lower bound
max_Lower_value=np.where(df['sepal_width']<= Lower)
print(max_Lower_value)
```

```python
min_upper_value=np.where(df['sepal_width']>= Upper)
print(min_upper_value)
```

```python
df.drop(max_Lower_value[0],inplace=True)
df.drop(min_upper_value[0],inplace=True)
print("new Shape:",df.shape)
```
